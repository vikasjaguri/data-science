```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.datasets import load_iris # to load the iris dataset
from sklearn.cluster import KMeans
```

```python
x,y = load_iris(return_X_y=True)
x
```

```python
y
```

```python
# Elbow Method
sse = [] # sum of squared errors
for i in range(1,11):
    kmodel = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmodel.fit(x)
    sse.append(kmodel.inertia_)
```

```python
g = sns.lineplot(x=range(1,11), y=sse)
g.set(xlabel='Number of Clusters', ylabel='SSE', title='Elbow Method')
```

```python
# Build Kmeans cluster model
Kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
Kmeans.fit(x)
```

```python
# find the cluster centers
Kmeans.cluster_centers_
```

```python
# Predict the clusters
pred = Kmeans.fit_predict(x)
pred
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
data
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

# Creating the Clusters

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='magma')
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='^', color='black')
```
